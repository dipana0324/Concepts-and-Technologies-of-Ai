Exploratory Data Analysis Practice Exercise - 1.

```python

df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
df["fare_per_year"] = df["Fare"] / df["Age"]

high_fare_age = df[df["fare_per_year"] > 5]

high_fare_age_sorted = high_fare_age.sort_values(by="fare_per_year", ascending=False)

high_fare_age_sorted[["Name", "Fare", "Age", "fare_per_year"]].head()
```

```python
df["fare_per_class"] = df["Fare"] / df["Pclass"]

adult_males = df[(df["Age"] >= 18) & (df["Sex"] == "male")]

adult_males_sorted = adult_males.sort_values(by="fare_per_class", ascending=False)

adult_males_sorted[["Name", "Age", "Pclass", "fare_per_class"]].head()
```

Group-By Exercises

1. What percent of total fare revenue came from each passenger class?

```python
total_fare = df["Fare"].sum()

fare_by_class = df.groupby("Pclass")["Fare"].sum()

percent_share = (fare_by_class / total_fare) * 100

percent_share
```

2. Percent of passengers belonging to each age group (child/adult/senior)

```python
def age_group(age):
    if age < 18:
        return "child"
    elif age >= 65:
        return "senior"
    else:
        return "adult"

df["age_group"] = df["Age"].apply(age_group)

age_counts = df["age_group"].value_counts()

age_percent = (age_counts / len(df)) * 100

age_percent
```
