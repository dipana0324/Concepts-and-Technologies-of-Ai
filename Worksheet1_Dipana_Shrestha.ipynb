8.Create a 2d array with 1 on border and 0 inside.

```python
def border_ones(n_rows, n_cols):
    a = np.zeros((n_rows, n_cols), dtype=int)
    a[0, :] = 1
    a[-1, :] = 1
    a[:, 0] = 1
    a[:, -1] = 1
    return a

border = border_ones(6, 6)  # example 6x6
print("2.8) border (6x6):\n", border, "\n")
```

Solve the following system of Linear equation using Inverse Methods.
2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9
{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}
• Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function
of Numpy}

```python
A = np.array([[2, -3, 1],[1, -1, 2],[3,  1, -1]])

B = np.array([-1, -3, 9])
```

Using inverse method: X = A⁻¹B

```python
A_inv = np.linalg.inv(A)
X = A_inv.dot(B)

print("Solution (x,y,z) =", X)
```

Solve using np.linalg.solve (better method)

```python
X2 = np.linalg.solve(A, B)
print("Solution using linalg.solve:", X2)
```

```python

import time
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
np.random.seed(0)

print("Imports ready. NumPy version:", np.__version__)
```

Create data (lists and arrays)

```python
# Cell 2: create data
N = 1_000_000            # for element-wise and dot tests
print("N:", N)

# Python lists
list1 = list(range(N))
list2 = list(range(N))

# NumPy arrays (same contents)
arr1 = np.arange(N)
arr2 = np.arange(N)

print("Created lists and NumPy arrays of length", N)
```

Element-wise addition (Python lists)

```python
# Cell 3: element-wise addition using Python lists
start = time.perf_counter()
list_add = [list1[i] + list2[i] for i in range(N)]
end = time.perf_counter()
time_list_add = end - start

print(f"Python list element-wise addition time: {time_list_add:.4f} seconds")
```

Element-wise addition (NumPy)

```python
# Cell 4: element-wise addition using NumPy
start = time.perf_counter()
np_add = arr1 + arr2
end = time.perf_counter()
time_np_add = end - start

print(f"NumPy element-wise addition time: {time_np_add:.6f} seconds")
```

Element-wise multiplication (Python lists & NumPy)

```python
# Cell 5: element-wise multiplication - lists
start = time.perf_counter()
list_mul = [list1[i] * list2[i] for i in range(N)]
end = time.perf_counter()
time_list_mul = end - start
print(f"Python list element-wise multiplication time: {time_list_mul:.4f} seconds")

# NumPy multiplication
start = time.perf_counter()
np_mul = arr1 * arr2
end = time.perf_counter()
time_np_mul = end - start
print(f"NumPy element-wise multiplication time: {time_np_mul:.6f} seconds")
```

Dot product (Python lists vs NumPy)

```python
# Cell 6: dot product - lists (using sum of products)
start = time.perf_counter()
dot_list = sum(list1[i] * list2[i] for i in range(N))
end = time.perf_counter()
time_list_dot = end - start
print(f"Python list dot product time: {time_list_dot:.4f} seconds")
print("Dot product value (list):", dot_list)

# NumPy dot
start = time.perf_counter()
dot_np = np.dot(arr1, arr2)
end = time.perf_counter()
time_np_dot = end - start
print(f"NumPy dot product time: {time_np_dot:.6f} seconds")
print("Dot product value (NumPy):", dot_np)
```

Matrix multiplication: quick test first (200x200) to verify code

```python
# Cell 7: quick matrix-multiplication test (small) to verify implementation
M = 200   # small test to ensure code works quickly
A_small = [[1]*M for _ in range(M)]
B_small = [[1]*M for _ in range(M)]

start = time.perf_counter()
C_small = [[sum(A_small[i][k] * B_small[k][j] for k in range(M)) for j in range(M)]
           for i in range(M)]
end = time.perf_counter()
print(f"Python list matrix multiplication time for {M}x{M}: {end - start:.4f} s")

# NumPy small
A_np_small = np.ones((M, M))
B_np_small = np.ones((M, M))
start = time.perf_counter()
C_np_small = A_np_small.dot(B_np_small)
end = time.perf_counter()
print(f"NumPy matrix multiplication time for {M}x{M}: {end - start:.6f} s")

# quick correctness check
print("Check element equality for small test:", C_small[0][0], C_np_small[0,0])
```
